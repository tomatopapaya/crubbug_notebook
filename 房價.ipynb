```python
import requests
```

```python
def get_soup(url):
    try:
        resp=requests.get(url)
        if resp.status_code==200:
            resp.encoding='utf-8'
            
            return BeautifulSoup(resp.text,'lxml')
        
    except Exception as e:
        print(e)
        
    return None
```

### 各縣市房價負擔表
- 縣市 房貸負擔率 房價所得比 增幅

```python
trs=soup.find('div',class_="table_body").find_all('tr')
trs
```

```python
len(trs)
```

```python
tr=trs[2]
```

```python
tr.find_all('td')  
```

```python
datas=[]
for i,tr in enumerate(trs):
    if i<2:
        continue
    else:
        data=[]
    for td in tr.find_all('td'):
            print(td.text.strip(),end='\t')
            data.append(td.text.strip())
    print(data)    
    datas.append(data)

datas    
```

```python
datas
```

```python
title=soup.find('table').find('thead')
title
```

```python
len(title)
```

```python
titles=[]
for tr in title.find_all('tr'):
    for i,td in enumerate(tr.find_all('th')):
        if i<0:
            continue
        else:
            print(td.text.strip(),end='\t')
            titles.append(td.text.strip())
    print()
    
titles[5:]
```

```python
columns=titles[5:]
```

```python
columns.insert(0,'縣市')
columns
```

```python
import pandas as pd
```

```python
datas
```

```python
df=pd.DataFrame(datas,columns=['縣市', '房貸負擔率(%)', '房貸負擔率(百分點)季變動值', '房貸負擔率(百分點)年變動值', '房價所得比(倍)', '房價所得比(倍)季變動值', '房價所得比(倍)年變動值'])
df
```
