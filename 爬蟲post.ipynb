### 組成函式

```python
def get_ibon_data(city):
    api_url='https://www.ibon.com.tw/retail_inquiry_ajax.aspx'
    form_data={
        'strTargetField': 'COUNTY',
        'strKeyWords': city
    }
    resp=requests.post(api_url,data=form_data)
    soup=BeautifulSoup(resp.text,'lxml')
    trs=soup.find('table',class_="font16").find_all('tr')
    
    datas=[]
    
    for tr in trs:
        datas.append([td.text.strip() for td in tr.find_all('td')])
    #for td in tr.find_all('td'):
    #datas.append(td.text.strip())
    
    return datas
```

```python
get_ibon_data('台北市')
```

### 輸出CSV

```python
for city in cities:
    print(city)
    datas=get_ibon_data(cities[city])
    df=pd.DataFrame(datas[1:],columns=datas[0])
    df.set_index('店號',inplace=True)
    df.to_csv(f'{path}{city}_ibon_data.csv',encoding='utf-8-sig')
    print('儲存成功.')

print('done!')
```
