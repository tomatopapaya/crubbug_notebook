### mytool 函數內容

```python
def get_soup(url):
    try:
        resp=requests.get(url)
        if resp.status_code==200:
            resp.encoding='utf-8'
            
            return BeautifulSoup(resp.text,'lxml')
        
    except Exception as e:
        print(e)
        
    return None
```
