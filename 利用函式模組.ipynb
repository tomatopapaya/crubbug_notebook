```python
datas=[]
for i,tr in enumerate(trs):
    data=[]
    for j,td in enumerate(tr.find_all('div',class_="td")):     
        # title
        if i==1 and j==3:
            print(td.h2.text.strip())
            data.append(td.h2.text.strip())
        else:
            print(td.text.strip(),end='\t')
            data.append(td.text.strip())   
                
        # url    
        if i==0 and j==3:
            data.append('url')
        elif j==3:
            data.append(td.find('a').get('href'))
   
            
    datas.append(data)
    
    print()
```

### 抓取固定td位置

```python
trs=soup.find('div',class_="rank_list table rankstyle1").find_all('div',class_="tr")
```

```python
datas=[]
for i,tr in enumerate(trs[1:]): #避開第0列
    #取得所有td元素
    tds=[td for td in tr.find_all('div',class_="td")]  
    #排行
    rank=tds[0].text.strip()
    #第一名特別取h2 
    if i==0:
        title=tds[3].h2.text.strip()
    else:
        title=tds[3].text.strip()
    #連結    
    link=tds[3].find('a').get('href')    
    
    print(rank,title,link)           
    
    datas.append([rank,title,link])

datas
```

```python
pd.DataFrame(datas,columns=['rank','title','url'])
```
